```python
# Concatenate the three dataframes vertically
combined_df = pd.concat([all_year_heart_df, all_year_lung_df, all_year_diabetes_df])

# Group by "year_of_death" and calculate the total number of deaths for each year for each dataframe
grouped_heart = all_year_heart_df.groupby("year_of_death").size()
grouped_lung = all_year_lung_df.groupby("year_of_death").size()
grouped_diabetes = all_year_diabetes_df.groupby("year_of_death").size()

# Create a DataFrame with all years from 2005 to 2015
all_years = pd.DataFrame(index=range(2005, 2016))

# Give names to the Series
grouped_heart.name = 'Heart Deaths'
grouped_lung.name = 'Lung Deaths'
grouped_diabetes.name = 'Diabetes Deaths'

# Merge the grouped data with the DataFrame containing all years for each dataframe
merged_heart = all_years.merge(grouped_heart, left_index=True, right_index=True, how='left').fillna(0)
merged_lung = all_years.merge(grouped_lung, left_index=True, right_index=True, how='left').fillna(0)
merged_diabetes = all_years.merge(grouped_diabetes, left_index=True, right_index=True, how='left').fillna(0)

# Plot the line chart for each dataframe
plt.figure(figsize=(12, 6))
plt.plot(merged_heart.index, merged_heart['Heart Deaths'], marker='o', label='Heart Deaths')
plt.plot(merged_lung.index, merged_lung['Lung Deaths'], marker='o', label='Lung Deaths')
plt.plot(merged_diabetes.index, merged_diabetes['Diabetes Deaths'], marker='o', label='Diabetes Deaths')

plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths by Year for Top 3 Diseases')
plt.legend()

# plt.savefig("output_allyear/Fig4.png")
plt.show()
```

```python
# Concatenate the three dataframes vertically
combined_df = pd.concat([all_year_heart_df, all_year_lung_df, all_year_diabetes_df])

# Grouping the entries in the "358_cause_recode_text" column
def categorize_disease(entry):
    if entry in ['All other forms of chronic ischemic heart disease (I20,I25.1-I25.9)', 'Acute myocardial infarction (I21-I22)', 'Atherosclerotic cardiovascular disease, so described (I25.0)', 'Congestive heart failure (I50.0)']:
        return 'Heart Disease'
    elif entry in ['Of trachea, bronchus and lung (C33-C34)', 'Other chronic obstructive pulmonary disease (J44)']:
        return 'Lung Disease'
    elif entry == 'Diabetes mellitus (E10-E14)':
        return 'Diabetes'
    else:
        return 'Others'

combined_df['Disease Group'] = combined_df['358_cause_recode_text'].apply(categorize_disease)

# Get unique race entries from the "race_text" column
unique_races = combined_df['race_text'].unique()

# Iterate over each unique race
for race in unique_races:
    # Filter the data for the current race
    race_data = combined_df[combined_df['race_text'] == race]
    
    # Count the occurrences of each cause of death for the current race and disease group
    cause_counts = race_data['Disease Group'].value_counts()
    
    # Create a pie plot for the current race and disease group
    plt.figure(figsize=(8, 8))
    plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Causes of Death Distribution for {race}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.savefig(f"Visuals_allyear/allyearFig_{race}.png")

    plt.show()
```
